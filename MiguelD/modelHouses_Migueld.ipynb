```python
def correlation_matrix(df):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm

    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet', 50)
    cax = ax1.imshow(df.corr(), interpolation="nearest", cmap=cmap)
    ax1.grid(True)
    plt.title('Correlation Table')
    labels=['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
    ax1.set_xticklabels(labels,fontsize=12)
    ax1.set_yticklabels(labels,fontsize=12)
    # Add colorbar, make sure to specify tick locations to match desired ticklabels
    fig.colorbar(cax, ticks=[.75,.1,.85,.90,.95,1])
    plt.show()

correlation_matrix(df)
```

```python
#Modeling
y = df[['price']]
X = df[['sqft_living', 'grade', 'sqft_above', 'sqft_living15', 'bedrooms', 'floors', 'waterfront', 'sqft_basement','view','yr_renovated', 'sqft_lot', 'yr_built']]
```

```python
# Cargar modelo
model = sm.OLS(y, X).fit()

# Hacer predicciones
predictions = model.predict(X)

# Resumen del modelo
model.summary()
```

```python
plt.scatter(y, predictions)
```
