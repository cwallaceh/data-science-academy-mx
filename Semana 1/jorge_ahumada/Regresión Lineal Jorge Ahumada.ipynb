# Data Preparation

```python
# Elimino las variables ID y DATE para quedarme solo con las variables que son numéricas.
del df['id']
del df['date']
```

```python
df.head()
```

```python
#Correlacion entre las variables que utilizaré en mi modelo

df[['price', 'sqft_living', 'sqft_lot', 'waterfront', 'view']].corr()
```

```python
# Genero una nueva variable con la interseccion pie cuadrado habitable con frente al agua
df['living_water'] = df['sqft_living'] * df['waterfront']
# Genero una nueva variable con la intersecion pie cuadrado del lote con frente al agua
df['lot_water'] = df['sqft_lot'] * df['waterfront']
```

```python
df[['sqft_living', 'sqft_lot','waterfront','living_water', 'lot_water' ]] .head()
```

# Modeling

Yo quiero saber cuando aumenta más un pie cuadrado de construcción frente al mar con respecto a no tener esta vista.

```python
# Selecciono mis variables
y = df.pop('price')
x = df[['sqft_living', 'sqft_lot','waterfront','living_water', 'lot_water' ]]
```

```python
# Cargar modelo
model = sm.OLS(y, x).fit()

# Resumen del modelo
model.summary()
```

# Evaluation

Se encontró lo siguiente:
1. Caeteris Paribus; al aumentar un pie cuadrado habitable en la casa a comprar, se incrementa el precio de la vivienda en 259.72 dólares.
2. Caeteris Paribus; al aumentar un pie cuadrado en el terreno de la casa a comprar, el precio de la vivienda disminuye -0.2536 dólares.
3. Caeteris Paribus; al aumentar un pie cuadrado habitable de la casa a comprar y que esta esté frente al mar, se incrementa el precio de la vivienda 354.35 dólares más con respecto a las que no están frente al mar.
4. Caeteris Paribus; al aumentar un pie cuadrado de terreno de la casa a comprar y que esté frente al mar, el precio de la vivienda disminuye 1.35 dólares con respecto a las que no están frente al mar.

Asimismo, es importante resaltar que se tiene una R cuadrada ajustada de .857 y todos los coeficientes son estadisticamente significativos.
