# Business Understanding
This dataset contains house sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015.

We need to forecast the price of the houses given certains conditions.

# Data Preparation

```python
del df['id']
del df['date']
```

```python
df['tota_sqft'] = df['sqft_basement'] + df['sqft_above']
```

```python
df.head()
```

# Modeling

```python
# Regresión lineal 1 variable
# Selección de variables
y = df[['price']]
X = df[['sqft_living', 'bedrooms', 'bathrooms', 'grade', 'yr_renovated', 'view', 'yr_built']]
```

```python
# Cargar modelo
model = sm.OLS(y, X).fit()

# Hacer predicciones
predictions = model.predict(X)

# Resumen del modelo
model.summary()
```
